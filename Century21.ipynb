```python
import requests
import pandas
from bs4 import BeautifulSoup

d = pandas.DataFrame(columns=['Price', 'Address', '# of Beds', '# of Baths', 'Size', 'Lot Size'])

r = requests.get("http://pyclass.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/", 
                 headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
c = r.content

soup = BeautifulSoup(c, "html.parser")

all = soup.find_all('div', {"class": "propertyRow"})
```

```python
base_url = "http://pyclass.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/#t=0&s="
for page in range(0, 30, 10):
    r=requests.get(base_url+str(page)+'.html', 
                 headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
    c = r.content

    soup = BeautifulSoup(c, "html.parser")

    all = soup.find_all('div', {"class": "propertyRow"})
    for item in all:
        price = item.find('h4', {"class": 'propPrice'}).text.replace("\n", "").replace(" ", "")
        add1 = item.find_all('span', {"class": 'propAddressCollapse'})[0].text
        add2 = item.find_all('span', {"class": 'propAddressCollapse'})[1].text
        fullAdd = add1 + add2
        try:
            noOfBed = item.find('span', {"class": 'infoBed'}).find('b').text
        except:
            noOfBed = 'NA'
        try:
            noOfBath = item.find('span', {"class": 'infoValueFullBath'}).find('b').text
        except:
            noOfBath = 'NA'
        try:
            size = item.find('span', {"class": 'infoSqFt'}).find('b').text + ' Sq. Ft'
        except:
            size = 'NA'

        lotSize = 'NA'
        for col in item.find_all('div', {'class': 'columnGroup'}):
            for feature_group, feature_name in zip(col.find_all('span', {"class":'featureGroup'}), col.find_all('span', {"class":'featureName'})):
                if 'Lot Size' in feature_group.text:
                    lotSize = feature_name.text

        d = d.append({"Price": price, "Address": fullAdd, "# of Beds": noOfBed, "# of Baths": noOfBath, "Size": size, "Lot Size": lotSize}, ignore_index=True)

    d.to_csv("Data.csv")
```
